```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
!wget "https://raw.githubusercontent.com/me-t1me/DIP_assignment_2/master/xray.png"
```

```python
# Load the original X-ray image
img = cv2.imread('xray.png', cv2.IMREAD_GRAYSCALE)

# Define the JPEG compression quality (0-100)
quality = 50

# Define the JPEG compression parameters
params = [int(cv2.IMWRITE_JPEG_QUALITY), quality]

# Compress the image using JPEG
retval, compressed = cv2.imencode('.jpg', img, params)

# Get the size of the compressed image
compressed_size = len(compressed)

# Decompress the image
decompressed = cv2.imdecode(compressed, cv2.IMREAD_GRAYSCALE)

# Calculate the compression ratio
compression_ratio = (img.shape[0] * img.shape[1]) / compressed_size

# Calculate the SNR and RMSE values between the decompressed and original images
snr = cv2.PSNR(img, decompressed)
rmse = cv2.norm(img, decompressed, cv2.NORM_L2) / (img.shape[0] * img.shape[1]) ** 0.5

# Print the compression ratio, SNR, and RMSE values
print('Compression ratio:', compression_ratio)
print('SNR:', snr)
print('RMSE:', rmse)
```

```python
cv2_imshow(img)
cv2_imshow(decompressed)
```
